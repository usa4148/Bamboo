```python
#
# ETL Project Group 6 Team Bamboo
#   Dan C.
#

import pandas as pd
file = 'Resources/location-identifier-database-202103.csv'
icao_df = pd.read_csv(file, encoding="ISO-8859-1",low_memory=False)

usa_icao_df = icao_df.loc[icao_df['country3'] == "USA"] # Select just US stationis
new_icao_df = usa_icao_df.reset_index(drop=True)
new_icao_df.dropna(subset = ["icao"], inplace=True)

new_icao_df.reset_index(drop=True)
clean_icao_df = new_icao_df.drop(['country2','country','stn_key', 'status','national_id', 'wmo','wban','ghcn','special'],1)
clean_icao_df.reset_index(drop=True, inplace=True)

clean_icao_df = clean_icao_df.drop_duplicates(subset=['icao'],keep='last')

df_out_file = 'Resources/icao.csv'
clean_icao_df.to_csv(df_out_file, index = True, header=True)
print (clean_icao_df)
```
